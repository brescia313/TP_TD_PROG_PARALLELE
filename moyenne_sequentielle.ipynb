```python
import pandas as pd

# Charger le fichier CSV généré
df = pd.read_csv("notes_etudiants.csv")

# Fonction séquentielle pour calculer la moyenne pondérée
def moyenne_seq(m, p, a):
    return (m*5 + p*4 + a*2) / 11

# Application séquentielle sur toutes les lignes
df["Moyenne_seq"] = df.apply(lambda row: moyenne_seq(row["Maths"], row["Physique"], row["Anglais"]), axis=1)

# Vérification : comparer avec les premières lignes
print(df.head(10))
```
