```python
import numpy as np
import pandas as pd

# Nombre d'étudiants
N = 1_000_000  

# Génération aléatoire des notes entre 0 et 20
maths = np.random.randint(0, 21, N)
physique = np.random.randint(0, 21, N)
anglais = np.random.randint(0, 21, N)

# Génération d'identifiants uniques
ids = np.arange(1, N+1)

# Listes de prénoms et noms fictifs
prenoms = ["Jean", "Marie", "Ali", "Fatou", "Moussa", "Awa", "Ibrahima", "Sophie", "Cheikh", "Ndeye"]
noms = ["Dupont", "Ndiaye", "Sow", "Diop", "Kane", "Ba", "Fall", "Diallo", "Ngom", "Mbengue"]

# Attribution aléatoire
prenom_random = np.random.choice(prenoms, N)
nom_random = np.random.choice(noms, N)

# Calcul de la moyenne pondérée
moyenne = (maths*5 + physique*4 + anglais*2) / 11

# Création du DataFrame
df = pd.DataFrame({
    "ID": ids,
    "Prenom": prenom_random,
    "Nom": nom_random,
    "Maths": maths,
    "Physique": physique,
    "Anglais": anglais,
    "Moyenne": moyenne
})

# Sauvegarde dans un fichier CSV
df.to_csv("notes_etudiants.csv", index=False)

# Affichage d’un extrait pour vérifier
print("Fichier CSV généré avec", N, "étudiants (ID, prénom, nom, notes, moyenne).")
print(df.head(10))  # affiche les 10 premières lignes
```
