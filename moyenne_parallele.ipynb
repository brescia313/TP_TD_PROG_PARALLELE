```python
import pandas as pd
import numpy as np
from numba import njit, prange

# Charger le fichier CSV
df = pd.read_csv("notes_etudiants.csv")

# Conversion des colonnes en tableaux NumPy
maths = df["Maths"].to_numpy()
physique = df["Physique"].to_numpy()
anglais = df["Anglais"].to_numpy()

# Coefficients
coef_maths, coef_physique, coef_anglais = 5, 4, 2
total_coef = coef_maths + coef_physique + coef_anglais

# Fonction parallèle avec Numba
@njit(parallel=True)
def moyenne_parallel(maths, physique, anglais):
    N = len(maths)
    moyennes = np.empty(N, dtype=np.float64)
    for i in prange(N):  # prange = boucle parallèle
        moyennes[i] = (maths[i]*coef_maths + physique[i]*coef_physique + anglais[i]*coef_anglais) / total_coef
    return moyennes

# Calcul parallèle
df["Moyenne_parallel"] = moyenne_parallel(maths, physique, anglais)

# Vérification sur un extrait
print(df.head(10))
```
